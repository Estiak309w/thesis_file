```python
def find_prefix(prefix,word):
    a = word[len(word)-len(prefix):]
    if(a==prefix):
        is_prefix =  True
    else:
        is_prefix = False
    return is_prefix

def stem(word):
    #dic ={"ই": ['চ্ছি','চ্ছিল','ছি','চ্ছ','চ্ছে','চ্ছেন','তেছি','তেছ','তেছে','তেছেন','য়েছিল','েয়েছিল ','েছিল'],    "": ['ের','টিকে','র','দের','েন','বেন','তে','েও','ে','ও','গুলো','গুলোর','েরগুলো','েরগুলোর','টা','টি','টির','কে','তে','রা','ও','তো','েছি','েছিল','েছে','ছ','ই','েছিলেন','েন','িস','ছিস','ছিলি','টির','েরটা','েরটার','েরগুলোর','ভাবে']} 
    dic ={"য়": ['চ্ছি','চ্ছিল','চ্ছ','চ্ছে','চ্ছেন'],    "": ['গুলো','গুলোর','েরগুলো','েরগুলোর','গুলা','গুলাকে','টা','টি','টির','টিকে','েরটা','ের','র','েরটার','কে','তে','রা','ও','তো','েছি','েছিল','েছে','েছিলেন','ছিস','ছিলি','ভাবে','তেছি','তেছ','তেছে','তেছেন','য়েছিল','েয়েছিল ','েছিল']} 
    new_word=word
    for key,value in dic.items():
        for i in range(len(value)):
            prefix = dic[key][i]
            if (find_prefix(prefix,word)==True):
                #print(find_prefix(prefix))
                #print(key)
                new_word = word.replace(prefix, key)
                return(new_word)
    if(new_word==word):
        return(word)
```

```python
sentence_1 = "ছবিটিকে ব্যান করা হয়েছে "

def stemmer(sentence):
    new_sentence = []
    words = sentence.split()
    for word in words:
        #print(word)
        a = stem(word)
        #print(a)
        new_sentence.append(a)
    #print(new_sentence)

    print(" ".join(new_sentence))
    sent = " ".join(new_sentence)
    return (sent)
    

print(stemmer(sentence_1))
```

```python
with open("datafile/abusive_religious.txt", "r",encoding="utf-8") as f:
    for line in f:
        sent = stemmer(line)
        with open("test.txt", "w",encoding="utf-8") as myfile:
            myfile.write((sent + '\n'))
            myfile.close()


```
