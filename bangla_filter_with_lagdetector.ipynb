```python
from langdetect import detect
```

```python
def filter(line):
    sentence=[]
    words = line.split()
    for word in words:
        try:
            if(detect(word) =='bn'):
                sentence.append(word)
                
        except :
            pass   
        
        
    return (" ".join(sentence))

import re
def digit_remover(text):
    text = re.sub(r'[\d]','',text)
    text = re.sub(r'[?\!\t\тАЫ\тАЩ\(\)\.\ред]+','',text)
    text = re.sub(r'[/\-\,]',' ',text)
    text = text.rstrip("\n\r")
    return text


a = filter("english is a ржЖржкржирж╛ржжрзЗрж░ржХрзЗ ржЖржорж╛ржжрзЗрж░ ржХржХрзНрж╕ржмрж╛ржЬрж╛рж░ ")
print(a)
```

```python
with open("raw_comment.txt", "r",encoding="utf-8") as f:
    for line in f:
        sent = filter(line)
        sent = digit_remover(sent)
        with open("test.txt", "w",encoding="utf-8") as myfile:
            myfile.write((sent))
            if(len(sent)!=0):
                print(sent)
            #myfile.close()
```
